```python
import requests
from bs4 import BeautifulSoup
import csv

url = "https://itc.gymkhana.iitb.ac.in/wncc/soc/"

# Send HTTP GET request and retrieve the HTML content
response = requests.get(url)
html_content = response.content

# Parse the HTML
soup = BeautifulSoup(html_content, "html.parser")
                                
# Create a CSV file
csv_file = open("project_details.csv", "w", newline="")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Project Name", "Category"])

# Find all the project elements
Project_elements = soup.find_all("div", class_="col-lg-4 col-6 mb-4 shuffle-item")

# Iterate over the project elements and extract the project name and category
for Project_element in Project_elements:
    project_name = Project_element.find("p", class_="lead text-center font-weight-bold text-dark").text.strip()
    project_category = Project_element["data-groups"]
    csv_writer.writerow([project_name, project_category])

# Close the CSV file
csv_file.close()

# Print the CSV file contents
with open("project_details.csv", "r") as file:
    csv_data = file.read()

print(csv_data) 
```
